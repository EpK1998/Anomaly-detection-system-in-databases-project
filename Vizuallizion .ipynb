```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#open all data and make a DF
df_train=pd.read_csv("TrainDataFrame_After_Cleaning.csv")
test_data = pd.read_csv('TestDataFrame_After_Cleaning.csv')
```

```python
#pca for the train data set
from sklearn.decomposition import PCA
n_components = 3  # Number of principal components for 3D visualization
pca = PCA(n_components=3)
df_train=df_train.sort_index()
X_pca = pca.fit_transform(df_train)
# Plot 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], marker='o', alpha=0.6)

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3D PCA Visualization')

plt.show()
```

```python
#pca for test data
from sklearn.decomposition import PCA
n_components = 3  # Number of principal components for 3D visualization
pca = PCA(n_components=3)
test_data=test_data.sort_index()
X_pca_test = pca.fit_transform(test_data)

```

```python

# Plot 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_pca_test[:, 0], X_pca_test[:, 1], X_pca_test[:, 2], marker='o', alpha=0.6)

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3D PCA Visualization')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
corr_matrix = df_train.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import seaborn as sns

# Select a subset of features for clarity
subset_features = ['FLOWCOOLPRESSURE', 'IONGAUGEPRESSURE', 'ETCHAUXSOURCETIMER', 'ACTUALSTEPDURATION', 'recipe']
sns.pairplot(df_train[subset_features])
plt.suptitle('Scatter Plot Matrix', y=1.02)
plt.show()
```

```python
#combintions of the colomns correlitions
sns.clustermap(df_train.corr(), annot=True, cmap='coolwarm', figsize=(10, 10))
plt.title('Cluster Map of Features')
plt.show()
```

```python
#target colomn behavior 
from pandas.plotting import lag_plot

plt.figure(figsize=(10, 6))
lag_plot(df_train['FLOWCOOLPRESSURE'])
plt.title('Lag Plot of FLOWCOOLPRESSURE')
plt.show()
```
