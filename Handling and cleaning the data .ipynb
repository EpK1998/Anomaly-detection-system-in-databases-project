```python
#remove outliers if 30% of features is outliers
def drop_outliers(df, threshold=0.3):
    # Calculate Q1, Q3, and IQR for each column
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1

    # Identify outliers for each column
    outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))

    # Count the number of outliers in each row
    outlier_counts = outliers.sum(axis=1)

    # Identify rows with outlier percentage more than the threshold
    rows_to_drop = df.index[outlier_counts / len(df.columns) > threshold]
    df = df.drop(rows_to_drop)

    return df

df = drop_outliers(df, threshold=0.4)
test_data=drop_outliers(test_data, threshold=0.4)
```
